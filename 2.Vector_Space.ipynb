# Vector Space

## Linear Combinations

Suppose we have vectors $v_{1}, v_{2}, ..., v_{n}$. Then $\alpha_{1}v_{1} + \alpha_{2}v_{2} + ... + \alpha_{n}v_{n}$ is called the linear combinations of the vectors and $\alpha_{1}, \alpha_{2}, ..., \alpha_{n}$ are called the coefficients of the vectors. For example, some linear combinations of the vectors $\langle 1, 3 \rangle$ and $\langle 2, 2 \rangle$ are:

$$
2\langle 1, 3 \rangle + 3\langle 2, 2 \rangle\\
3\langle 1, 3 \rangle + 1\langle 2, 2 \rangle\\
0\langle 1, 3 \rangle + 0\langle 2, 2 \rangle\\
$$
A linear combination is said to be trivial if all the coefficients are zero. The question is can we devise an algorithm to express a vector as a linear combination of the given vectors? Will the solution be unique?

## Span

The set of all linear combinations of vectors $v_{1}, v_{2}, ..., v_{n}$ is called the span of the vectors and represented by $\{v_{1}, v_{2}, ..., v_{n}\}$. How many vectors are there in the span of $\langle 2, 3 \rangle$ over $R$? There are infinite many vectors $\{\alpha \langle 2, 3 \rangle, \alpha \in R\}$, and they form a line through the origin.

1. Span of a single non-zero vector $v$ = $\{\alpha v: \alpha \in R\}$ is a line through origin (One dimensional).
2. Span of a empty set is just the origin (Zero Dimensional).
3. Span of two vectors is all the points in the plane (Two dimensional).
4. Span of three vectors is planes in three dimensions.

The span of $k$ vectors is not always k-dimensional. For example, $\langle 0,0 \rangle$ is zero dimensional. Again, span of $\{[1,3],[2,6]\}$ is a line since second vector is just a linear combination of the first. So, given $k$ vectors, how can we predict its dimensions?

Let's consider a challange-response password authentication system where $x$ is the password, $a_{1}, a_{2}, ..., a_{m}$ are challanges by server, and $\beta_{1}, \beta_{2}, ..., \beta_{m}$ are the responses by the user. Suppose, response is created by the dot product of challange and password. We can represent this by following linear equations:

$$
a_{1}x = \beta_{1}\\
a_{2}x = \beta_{2}\\
.\\
.\\
.\\
a_{m}x = \beta_{m}\\
$$

Now, an evesdropper can see the challanges and the responses. Can he figure out the response for future challanges without knowing the password? Any vectors $a$ in the span of the challange vectors can be represented as $a = \alpha_{1}a_{1} + ... + \alpha_{m}a_{m}$. Then,

$$
a.x = (\alpha_{1}a_{1} + ... + \alpha_{m}a_{m})x\\
= \alpha_{1}a_{1}x + ... + \alpha_{m}a_{m}x\\
= \alpha_{1}(a_{1}x) + ... + \alpha_{m}(a_{m}x)\\
= \alpha_{1}\beta_{1} + ... + \alpha_{m}\beta_{m}
$$

So, an evesdropper can compute any response that lie in the span of the extracted challanges. Can he derive responses for challanges that do not lie in the span of challange vectors?

## Generators

If $v_{1}, v_{2}, ..., v_{n}$ are vectors such that $V = \{v_{1}, v_{2}, ..., v_{n}\}$ or $V$ is the span of the vectors, then we say that,
1. $\{v_{1}, v_{2}, ..., v_{n}\}$ is the generating set of $V$.
2. We say the vectors $v_{1}, v_{2}, ..., v_{n}$ are to be generators of $V$.

To say that $\{[3,0,0],[0,2,0],[0,0,1]\}$ is a generating set of $R^{3}$, we need to show that,
1. Every linear combinations of the generating set is a vector in $R^{3}$. It is easy to proof since any combination will have 3 elements and so will be in $R^{3}$.
2. Every vector in $R^{3}$ is a linear combination of the vectors in the generating set. Any vector in $R^{3}$, $\langle x,y,z \rangle$ can be written as the linear combination of the generators, for example, $(x/3)[3,0,0] + (y/2)[0,2,0] + z[0,0,1]$.

Suppose we claim that $\{[1,0,0],[1,1,0],[1,1,1]\}$ is also a generating set of $R^{3}$. We already know that $\{[3,0,0],[0,2,0],[0,0,1]\}$ is a generating set of $R^{3}$. We just need to proof that $\{[3,0,0],[0,2,0],[0,0,1]\}$ is in the span of $\{[1,0,0],[1,1,0],[1,1,1]\}$ by writing it as a linear combination of the new set. For example, $[3,0,0] = 3[1,0,0] + 0[1,1,0] + 0[1,1,1]$. We can derive for other vectors in the old generating set in similar way. We then can substitue $[3,0,0],[0,2,0],[0,0,1]$ in the $(x/3)[3,0,0] + (y/2)[0,2,0] + z[0,0,1]$ in terms of new vector to proof that the new vector is also a generating set of $R^{3}.

We can think of the simplest generators for $R^{3}$ as $\{[1,0,0],[0,1,0],[0,0,1]\}$ so that we can simply represent any vector $\langle x,y,z \rangle$ as $x[1,0,0] + y[0,1,0] + z[0,0,1]$. These generators are called *standard generators*, and often represented as $e_{1},e_{2},e_{3}$.

## Vector Space & Subspace

Geometric objects like lines and planes can be represented by 2 ways:
1. Span of some vectors.
2. Solution set of some system of linear equations with zero in the right hand side.

Suppose, line $l_{1}$ is represented as the span of some vector $v_{1}$, and $l_{2}$ as the span of another vector $v_{2}$. Now, if we want to find the plane that contains both of them. That plane will be the span of $\{v_{1}, v_{2}\}$. Again, suppose two planes are represented as solution sets of linear equations like $\{[x,y,z]: [4,-1,1].[x,y,z] = 0 \}$ and $\{[x,y,z]: [0,1,1].[x,y,z] = 0 \}$. You want know about the line where those planes intersect. One way to do that will be to find variables $\langle x,y,z \rangle$ for which both equations are satisfied.

Subset of $R^{D}$ satisfies following 3 properties:
1. (V1) Subset contains the zero vector. For example, span of $v_{1},..., v_{n}$ always contains $0 \times v_{1}+ ... + 0 \times v_{n} = 0$. Similarly, for linear equation representaion, if the vector is zero vector, the equation will be satisfied regardless of the coefficients. For example, $\{[x,y,z]: [4,-1,1].[x,y,z] = 0 \}$ always has a solution when $[x,y,z] = [0,0,0]$. So, zero vector is in the subset of the solution set.
2. (V2) If a subset contains $v$, then it must contain $\alpha v$ for every scalar $\alpha$.
3. (V3) If a subset contains $u$ and $v$, it must contain $u + v$.

Any subset $\nu$ of $R^{D}$ satisfying the above 3 properties are called vector space. If $\upsilon$ is also vector space and subset of $\nu$, it is called a subspace of $\nu$.

## Convex & Affine Hull

Just like convex combinations of line segments, we can take the convex combinations of vectors $\langle v_{1},v_{2},...,v_{n} \rangle$ like $\alpha_{1}v_{1} + \alpha_{2}v_{2} + ... + \alpha_{n}v_{n}$, where the coefficients are nonnegative and sums to 1.

1. Convex hull of a single vector is point.
2. Convex hull of two vectors is a line segment.
3. Convex hull of three vectors is a triangle.

Similarly, we can take the affine combinations of the vectors. The set of all affine combinations of some vectors is called the affine hull of those vectors. Affine hull of $u_{1},u_{2},...,u_{n}$ can be written as $\{u_{1} + Span\{u_{2}-u_{1},u_{3}-u_{1},...,u_{n}-u_{1}\}\}$

**Solution set of a system of linear equation is either empty or an affine space.**

A linear equation of the form $a.x = 0$ is called a homogeneous linear equation. A system of homogeneous linear equations is called a homogeneous linear system. Any linear system such as the following

$$
a_{1}x = \beta_{1}\\
a_{2}x = \beta_{2}\\
.\\
.\\
a_{m}x = \beta_{m}
$$

has a corresponding homogeneous linear system where all the entries in right hand side is zero, like the following


$$
a_{1}x = 0\\
a_{2}x = 0\\
.\\
.\\
a_{m}x = 0
$$
